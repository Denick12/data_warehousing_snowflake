```python
use database snowflake_learning_db;
create schema denick;
use schema denick;

-- How we create our models

-- DIMENSIONAL TABLES
-- create products dim table, customer dim, time dim, store dim, then create employee dim

-- FACTS TABLE
-- sales transaction facts table
-- inventory movement facts table
-- customer interactions facts table

create database retail_analytics;
create schema retail_data;

create table DIM_products (
product_id int primary key,
product_name varchar(50),
product_category varchar(100),
product_brand varchar(50),
product_price float
);

create table DIM_customer(
customer_id int primary key,
customer_full_name varchar (50),
customer_age int,
customer_gender varchar(100),
customer_location varchar(100),
customer_email varchar(100)
);

create table DIM_time(
date_id int primary key,
date date,
month int,
quarter int,
year int,
day_of_week varchar(100)
);

create table DIM_store(
store_id int primary key,
store_name varchar(100),
city varchar(100),
region varchar(100),
store_type varchar(100)
);

create table DIM_employees(
employee_id int primary key,
employee_full_name varchar(100),
employee_position varchar(100),
hire_date varchar(100),
store_id int,
foreign key (store_id) references DIM_store(store_id)
);


-- create table FACT_sales(
-- transaction_id int primary key,
-- date_id int,
-- product_id int,
-- customer_id int,
-- store_id int,
-- quantity_sold int,
-- total_amount float,
-- foreign key (date_id) references DIM_time(date_id),
-- foreign key (product_id) references DIM_products(product_id),
-- foreign key (customer_id) references DIM_customer(customer_id),
-- foreign key (store_id) references DIM_store(store_id)
-- );

CREATE TABLE FACT_sales (
    transaction_id INT PRIMARY KEY,
    date_id INT,
    product_id INT,
    customer_id INT,
    store_id INT,
    quantity_sold INT,
    total_amount FLOAT,
    FOREIGN KEY (date_id) REFERENCES DIM_time(date_id),
    FOREIGN KEY (product_id) REFERENCES DIM_products(product_id),
    FOREIGN KEY (customer_id) REFERENCES DIM_customer(customer_id),
    FOREIGN KEY (store_id) REFERENCES DIM_store(store_id)
);


-- create table FACT_inventory(
--     movement_id int primary key,
--     date_id int,
--     product_id int,
--     store_id int,
--     movement_type varchar(100),
--     quantity_change int,
--     foreign key (date_id) references DIM_time(date_id),
--     foreign key (product_id) references DIM_products(product_id),
--     foreign key (store_id) references DIM_store(store_id)
-- );

CREATE TABLE FACT_inventory (
    movement_id INT PRIMARY KEY,
    date_id INT,
    product_id INT,
    store_id INT,
    movement_type VARCHAR(100),
    quantity_change INT,
    FOREIGN KEY (date_id) REFERENCES DIM_time(date_id),
    FOREIGN KEY (product_id) REFERENCES DIM_products(product_id),
    FOREIGN KEY (store_id) REFERENCES DIM_store(store_id)
);


-- create table FACT_customer(
-- interaction_id int primary key,
-- date_id int,
-- customer_id int,
-- employee_id int,
-- interaction_type varchar(100),
-- feedback_score int,
-- foreign key (date_id) references DIM_time(date_id),
-- foreign key (customer_id) references DIM_customer(customer_id),
-- foreign key (employee_id) references DIM_employees(employee_id)
-- );

CREATE TABLE FACT_customer (
    interaction_id INT PRIMARY KEY,
    date_id INT,
    customer_id INT,
    employee_id INT,
    interaction_type VARCHAR(100),
    feedback_score INT,
    FOREIGN KEY (date_id) REFERENCES DIM_time(date_id),
    FOREIGN KEY (customer_id) REFERENCES DIM_customer(customer_id),
    FOREIGN KEY (employee_id) REFERENCES DIM_employees(employee_id)
);






```
